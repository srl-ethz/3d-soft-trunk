```python
import numpy as np
import pandas as pd
import ast
import re

from pydrake.math import RigidTransform, RollPitchYaw, RotationMatrix
from pydrake.common.eigen_geometry import Quaternion

import matplotlib.pyplot as plt
%matplotlib inline
np.set_printoptions(precision=5)
pd.set_option('display.float_format', lambda x: '%.5f' % x)
```

```python
def make_xform(x, y, z, qw, qx, qy, qz):
    xyz = np.array([x, y, z])
    q_wxyz_array = np.array([qw, qx, qy, qz])
    q_wxyz_array = q_wxyz_array / np.linalg.norm(q_wxyz_array)
    q_wxyz = Quaternion(w=q_wxyz_array[0], x=q_wxyz_array[1], y=q_wxyz_array[2], z=q_wxyz_array[3])
    return RigidTransform(quaternion=q_wxyz, p=xyz)

def quat_dist_xform(X_AB):
    return quat_dist(X_AB.rotation().ToQuaternion(), Quaternion())

def get_rpy(xform):
    """[summary]

    Args:
        xform ([type]): [description]

    Returns:
        [type]: [description]
    """
    return RollPitchYaw(xform.rotation()).vector()

def quat_dist(q1, q2):
    """[summary]

    Args:
        q1 ([type]): [description]
        q2 ([type]): [description]

    Returns:
        [type]: [description]
    """

    q1_dot_q2 = np.dot(q1.wxyz(), q2.wxyz())

    # https://math.stackexchange.com/a/90098

    # distance w/o trig functions
    # return np.sqrt(np.power(1 - q1_dot_q2 * q1_dot_q2, 2))

    # angular distance
    return np.arccos(2 * np.power(q1_dot_q2, 2) - 1)
```

```python
df_csv_fpath = "/src/3d-soft-trunk/examples_python/20211007-f2-p0to6bar-n50-v1-wait5secs.csv"
# df_csv_fpath = "/src/3d-soft-trunk/examples_python/20211007-f2-p0to6bar-n50-v1.csv"
df = pd.read_csv(df_csv_fpath)
df.head(5)
```

```python
finger_idx = 2
finger_name = "pinky"
wait_time_s = 3

_RE_COMBINE_WHITESPACE = re.compile(r"(?a:\s+)")
```

```python
df.P_desired = df.P_desired.map(ast.literal_eval)
df.P_desired = df.P_desired.map(lambda x: x[finger_idx])

df.P_measured = df.P_measured.map(ast.literal_eval)
df.P_measured = df.P_measured.map(lambda x: x[finger_idx])

df["X_BT_meas.xyz_qwxyz"] = df["X_BT_meas.xyz_qwxyz"].map(lambda x: _RE_COMBINE_WHITESPACE.sub(" ", x[1:-1].strip()))
df["X_BT_meas.xyz_qwxyz"] = df["X_BT_meas.xyz_qwxyz"].map(lambda x: "[" + _RE_COMBINE_WHITESPACE.sub(", ", x) + "]")
df["X_BT_meas.xyz_qwxyz"] = df["X_BT_meas.xyz_qwxyz"].map(ast.literal_eval)
# df["X_BT_meas.xyz_qwxyz"] = df["X_BT_meas.xyz_qwxyz"].map(lambda x: ast.literal_eval(", ".join(x.split())))

for idx, field in enumerate(["x", "y", "z", "qw", "qx", "qy", "qz"]):
    df[f"{field}_BT_meas"] = df["X_BT_meas.xyz_qwxyz"].map(lambda x: x[idx])

df.flex_sensor = df.flex_sensor.map(ast.literal_eval)
for idx in range(3):
    df[f"flex_sensor_{idx}"] = df.flex_sensor.map(lambda x: x[idx])
df.flex_sensor = df.flex_sensor.map(lambda x: x[finger_idx])
```

```python
df['X_BT'] = df.apply(lambda row_data: make_xform(x = row_data['x_BT_meas'], y = row_data['y_BT_meas'], z = row_data['z_BT_meas'], qw = row_data['qw_BT_meas'], qx = row_data['qx_BT_meas'], qy = row_data['qy_BT_meas'], qz = row_data['qz_BT_meas']), axis=1)
```

```python
df['rpy_BT'] = df.X_BT.map(get_rpy)
df['rot_r_BT'] = df.X_BT.map(lambda x: get_rpy(x)[0])
df['rot_p_BT'] = df.X_BT.map(lambda x: get_rpy(x)[1])
df['rot_y_BT'] = df.X_BT.map(lambda x: get_rpy(x)[2])
```

```python
df['quat_dist_recomputed'] = df.X_BT.map(quat_dist_xform)
```

```python
# P_measured vs sensor
fig = plt.figure(figsize=(10, 10))

for desired_P_val in df.P_desired.unique(): 
    plt.scatter(df.P_measured[df.P_desired == desired_P_val], df.flex_sensor[df.P_desired == desired_P_val], label=f"Pd = {desired_P_val} Pa")
plt.ylabel("Flex sensor reading")
plt.xlabel("Pressure measured (Pa)")
title = f"Measured pressure vs Flex sensor reading - {finger_name} - wait: {wait_time_s}s"
plt.title(title)
plt.legend()
plt.savefig("pressure_vs_sensor.jpg")
```

```python
# q_distance vs sensor
fig = plt.figure(figsize=(10, 10))
for desired_P_val in df.P_desired.unique(): 
    plt.scatter(df.quat_dist_recomputed[df.P_desired == desired_P_val], df.flex_sensor[df.P_desired == desired_P_val], label=f"Pd = {desired_P_val} Pa")
plt.ylabel("Flex sensor reading")
plt.xlabel("Quaternion distance from mocap (rad)")
title = f"Quaternion distance vs Flex sensor reading - {finger_name} - wait: {wait_time_s}s"
plt.title(title)
plt.legend()
plt.savefig("quat_vs_sensor.jpg")
```

```python
# pressure_vs_sensor
fig = plt.figure(figsize=(10, 10))
for desired_P_val in df.P_desired.unique():
    plt.scatter(df.P_measured[df.P_desired == desired_P_val], df.quat_dist_recomputed[df.P_desired == desired_P_val], label=f"Pd = {desired_P_val} Pa")
plt.ylabel("Quaternion distance from mocap (rad)")
plt.xlabel("Pressure measured (Pa)")
title = f"Pressure measured vs Quaternion distance - {finger_name} - wait: {wait_time_s}s"
plt.title(title)
plt.legend()
plt.savefig("pressure_vs_quat.jpg")
```
